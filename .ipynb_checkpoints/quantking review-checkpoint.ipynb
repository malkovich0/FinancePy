```python
stockCode = 'A005930'
listCategory = ['연간매출','연간영익','연간순익','매출총이익율','판관비율','영업활동현흐','총자산','자본총계','시총',
               'EV/EBITDA','부채비율','차입율']
dicNameConveted = {'연간매출':'매출액','연간영익':'영업이익','연간순익':'지배순이익','영업활동현흐':'영업활동현금흐름',
                  '자본총계':'지배자산'}
# listCategory=['총자산']
dfAll = pd.DataFrame()
for stockCode in df.index:
    try:
        dfStock = pd.DataFrame(columns = ['stock_code','period'])
        for Category in listCategory:
            listValue = list(df.loc[stockCode,[Category]])[::-1]
            listPeriod = [i[1][:5] for i in df.loc[stockCode,[Category]].index][::-1]
            listValueTemp = []
            listPeriodTemp = []
            for i in range(len(listPeriod)):
                if listPeriod[i][-1] == 'Q':
                    periodConverted = '20%s'%(listPeriod[i][0:2])+'%02d'%(int(listPeriod[i][3])*3)
                    if periodConverted not in listPeriodTemp:
                        listValueTemp.append(listValue[i])
                        listPeriodTemp.append(periodConverted)
                elif (listPeriod[i][-1] == '년') and (listPeriod[i] != "내년"):
                    for month in ['12','09','06','03']:
                        periodConverted = '20%02d'%int(listPeriod[i][:-1])+month
                        if periodConverted not in listPeriodTemp:
                            listValueTemp.append(listValue[i])
                            listPeriodTemp.append(periodConverted)
            if Category in dicNameConveted:
                Category = dicNameConveted[Category]

            df_stock_temp = pd.DataFrame({'period':listPeriodTemp,Category:listValueTemp})
            df_stock_temp['stock_code'] = stockCode[1:]
            dfStock = pd.merge(dfStock,df_stock_temp,on=['stock_code','period'],how='outer')
        dfStock['매출총이익'] = dfStock['매출액']*dfStock['매출총이익율']/100
        dfStock = dfStock[['stock_code','period','총자산','지배자산','매출액','매출총이익','영업이익','지배순이익','영업활동현금흐름',
                          'EV/EBITDA','부채비율','차입율']]

        dfAll = pd.concat([dfAll,dfStock],axis=0)
    except:
        print(stockCode)
dfAll
```
