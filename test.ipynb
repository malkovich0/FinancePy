```python
from pykrx import stock
kospi_list = stock.get_market_ticker_list("20210905",market='KOSPI')
kosdaq_list = stock.get_market_ticker_list("20210905",market='KOSDAQ')
import FinanceDataReader as fdr
krx_del_list_all = fdr.StockListing('KRX-DELISTING')['Symbol'].to_list()
krx_del_list = []
for code in krx_del_list_all:
    if code[0:5].isdigit():
        if len(code) == 6:
            krx_del_list.append(code)
krx_list = kospi_list + kosdaq_list + krx_del_list
```

```python
dictClose = {}
dictChange = {}
for stockCode in krx_list:
    if krx_list.index(stockCode) % 100 == 0:
        print(krx_list.index(stockCode))
    
    try:
        data = fdr.DataReader(stockCode)
        dictClose[stockCode] = data['Close']
        dictChange[stockCode] = data['Change']
    except:
        print(stockCode)
```

```python
import pickle

with open('Raw_Price/change_all_modi.pickle','wb') as handle:
    pickle.dump(dictChange, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('Raw_Price/price_all_modi.pickle','wb') as handle:
    pickle.dump(dictClose, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
